# Data transform &amp; split 

```
# import packages 
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
```

## Gold - Data transformation

```
# import data 
gold_data = pd.read_csv("silver_data_model.csv")
```

### One hot encode data - street (only run if encoded data is needed)!! 

```


def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    res = pd.concat([original_dataframe, dummies], axis=1)
    return(res)


# one hot encode the street 
gold_data = encode_and_bind(gold_data, 'street_name')


# one hot encode YEAR 
gold_data['year'] = gold_data['year'].astype(str)
gold_data = encode_and_bind(gold_data, 'year')
```

```
gold_data
```

```
# save dataset as csv 
gold_data.to_csv("gold_data_transformed.csv", index=False)
```

## Data split - train and test data

```
# gold_model_data = pd.read_csv("gold_data.csv")

gold_data_model = pd.read_csv("gold_data_transformed.csv")
```

```
gold_data_model.head()
```

```
# review data 
pd.DataFrame(gold_data_model.dtypes)
```

# Split data 

```
# train data 2019-2022.01
train_data_1 = gold_data_model.loc[(gold_data_model['year'] == 2022) & (gold_data_model['month'] == 1)]
train_data_2 = gold_data_model.loc[gold_data_model['year'] <= 2021]

train_data = pd.concat([train_data_1,train_data_2])

# test data 2022.02-2022.03: 
test_data_1 = gold_data_model.loc[(gold_data_model['year'] == 2022) & (gold_data_model['month'] == 2)]

test_data_2 = gold_data_model.loc[(gold_data_model['year'] == 2022) & (gold_data_model['month'] == 3)]

test_data = pd.concat([test_data_1,test_data_2])
```

```
test_data.describe()
```

```
train_data.head()
```

### remove columns

```
# drop col for train_val
train_data = train_data.drop(columns=[
    'street_name',
    'year',
    'sold_date']) 

# drop col for test
test_data = test_data.drop(columns=[
    'street_name',
    'year',
    'sold_date']) 
```

### Split train test

```
# split TRAIN and drop target aka. sold_price

X_train = train_data.drop(['sold_price'], axis= 1)
y_train = train_data['sold_price']
print(X_train.shape, y_train.shape)


# split TEST and drop target aka. sold_price

X_test = test_data.drop(['sold_price'], axis= 1)
y_test = test_data['sold_price']
print(X_test.shape, y_test.shape)
```

### Export data splits as .npy files 

```
X_train.to_csv("X_train.csv", index=False)
y_train.to_csv("y_train.csv", index=False)

X_test.to_csv("X_test.csv", index=False)
y_test.to_csv("y_test.csv", index=False)

#test_data.to_csv("test_data_2022_03.csv", index=False)
```

### Data stats 

```
train_len = len(X_train)
test_len = len(X_test)

print('train ratio:', train_len / (train_len+test_len)*100)
print('test ratio:', test_len / (train_len+test_len)*100)
```
